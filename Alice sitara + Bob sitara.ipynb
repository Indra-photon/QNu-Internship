```python
## This is the function to get the count value for a particular DLI value. 

import paramiko
import time
import json

def telnet_connection(x):
    # Start an interactive shell
    shell = client2.invoke_shell()

    # Wait for the shell to be ready
    time.sleep(1)

    # Clear the output buffer
    while shell.recv_ready():
        shell.recv(4096)

    # Send the Telnet command to localhost
    shell.send('telnet localhost 8000\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    while not output.endswith('Username: '):
        output += shell.recv(4096).decode()
    print(output)

    # Send the username
    shell.send('admin\n')
    time.sleep(1)

    # Wait for the password prompt
    output = ''
    while not output.endswith('Password: '):
        output += shell.recv(4096).decode()
    print(output)

    # Send the password
    shell.send('qunu123\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    while not output.endswith('qunu> '):
        output += shell.recv(4096).decode()
    print(output)

    # Send the command to set the float value
    shell.send('diagonise bem set SET_DLI_CNTRL\n')
    time.sleep(1)

    output = ''
    output += shell.recv(4096).decode()
    print(output)

    shell.send(x)
    time.sleep(1)

    output = ''
    output += shell.recv(4096).decode()
    print(output)

    # Wait for the command prompt
    #output = ''
    #while not output.endswith('qunu> '):
     #   output += shell.recv(4096).decode()
    #print(output)

    # Exit the Telnet session
    shell.send('exit\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    while not output.endswith('# '):
        output += shell.recv(4096).decode()
    print(output)

    shell.send('tail -f  /var/log/qunu_zynq_json\n')
    time.sleep(1)


    # Wait for the command prompt
    output = ''
    #while not output.endswith('# '):
    output += shell.recv(4096).decode()
    #print(output)

    # Split the output into lines
    lines = output.strip().split('\n')

    # Remove the first line if it matches "tail -f  /var/log/qunu_zynq_json"
    if lines[0].strip() == "tail -f  /var/log/qunu_zynq_json":
        lines = lines[1:]

    # Process the remaining lines
    for line in lines:
        line = line.strip()  # Remove leading/trailing whitespace
        if line:
            data = {}
            try:
                data = json.loads(line)
                set_value = x
                timestamp = data.get('ts')
                count1 = data.get('Click_Count1')
                count2 = data.get('Click_Count2')
                Extinction_ratio = abs(count2-count1)
                message = data.get('Message')
                data = {"Set Value": set_value, "Timestamp": timestamp, "Click_count1": count1, "Click_count2": count2,"Extinction ratio": 
                        Extinction_ratio, "Message": message }
            except json.JSONDecodeError as e:
                print(f"Error parsing JSON: {e}")
    return(data)
```

```python
# Here we are sweeping the DLI value and get the count. Then we will take the DLI value corresponding to maximum extinction ratio

import pandas as pd
set_value = []
#steps = input("Give the value of steps:") # Asking the user to give steps or you can set it to 0.05, or smaller than that if you want
initial_value = 0.1
while initial_value < 2.5:
    x = str(initial_value) + '\n'
    set_value.append(x)
    initial_value = initial_value + 0.05
#print(set_value)

# Calling the function for each set value and collect the counts
results = []
for a in set_value:
    result = telnet_connection(a) # call the function for each set_ value and collect the data
    results.append(result)

    
ex_ratio = []
set_val = []
for i in range(len(results)):
    ex_ratio.append(results[i]['Extinction ratio'])
    set_val.append(results[i]['Set Value'])
    
print(max(ex_ratio))
index = (ex_ratio.index(max(ex_ratio)))
#print(set_val)
print(set_val[index])
set_DLI_Value = set_val[index]   
    
# Post-processing the data in a excel sheet

#Post-processing
df = pd.DataFrame(results)  # Or you can print it in Excel as well
df
# Create an Excel writer
writer = pd.ExcelWriter('Data Count_for_DLI_Value.xlsx', engine='xlsxwriter')

# Write the DataFrame to the Excel sheet
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Add the plot image to the Excel sheet
workbook = writer.book
worksheet = writer.sheets['Sheet1']
worksheet.set_column('A:A', 20)
worksheet.set_column('B:B', 20)
worksheet.set_column('C:C', 20)
worksheet.set_column('D:D', 20)
worksheet.set_column('E:E', 20)
worksheet.set_column('F:F', 20)
worksheet.set_column('G:G', 20)
worksheet.set_column('H:H', 20)

cell_format = workbook.add_format({'bold': True, 'text_wrap': True})
writer.save()
```

```python
# This part is for cheking the standard deviation of DLI. We will receive the count1 and count2 with the set DLI value for 5 minutes
# and check the standard deviation for count1 or count2 (whichever corresponds to minimum of min(count1) or min(count2))

import time

count1 = []
count2 = []
DLI = str(set_DLI_Value) + '\n'

# Set the duration of the loop in seconds (5 minutes = 300 seconds)
loop_duration = 300

# Get the current time in seconds since the epoch (January 1, 1970)
start_time = time.time()

# Calculate the end time of the loop
end_time = start_time + loop_duration

# Run the loop until the current time reaches the end time
while time.time() < end_time:
    res = telnet_connection(DLI) # DLI value we will receive from the previous step 
    count1.append(res['Click_count1'])
    count2.append(res['Click_count2'])
    time.sleep(1)  # Optional: Add a 1-second delay between function calls to avoid CPU overload

```

```python
import statistics 

def calculate_standard_deviation(count1, count2): #count1 and count2 are list 
    if min(count1)<min(count2):
        std_dev = statistics.stdev(count1)
    else:
        std_dev = statistics.stdev(count2)
    return(std_dev)

standard_deviation = calculate_standard_deviation(count1,count2)
if standard_deviation < 200:
    print(f"standard deviation is {standard_deviation}, Passed! Proceed to the next stage")
else:
    print(f"standard deviation is {standard_deviation}, Failed, Do the DLI set again")
```
