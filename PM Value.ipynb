```python
import serial
import pandas as pd
import numpy as np
import time
import math
import pyvisa
from matplotlib import pyplot as plt
import xlsxwriter

def telnet_connection_PM(x):
    # Start an interactive shell
    shell = client1.invoke_shell()

    # Wait for the shell to be ready
    time.sleep(1)

    # Clear the output buffer
    while shell.recv_ready():
        shell.recv(4096)

    # Send the Telnet command to localhost
    shell.send('telnet localhost 8000\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    #while not output.endswith('Username: '):
    output += shell.recv(4096).decode()
    print(output)

    # Send the username
    shell.send('admin\n')
    time.sleep(1)

    # Wait for the password prompt
    output = ''
    #while not output.endswith('Password: '):
    output += shell.recv(4096).decode()
    print(output)

    # Send the password
    shell.send('****\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    #while not output.endswith('qunu> '):
    output += shell.recv(4096).decode()
    print(output)

    # Send the command to set the float value
    #shell.send('diagonise bem set SET_DLI_CNTRL\n')
    shell.send('diagonise aem set AAM_SUBSYS/PM_SPI_AXI\n')
    time.sleep(1)

    output = ''
    output += shell.recv(4096).decode()
    print(output)

    shell.send(x)
    time.sleep(1)

    output = ''
    output += shell.recv(4096).decode()
    print(output)

    # Wait for the command prompt
    #output = ''
    #while not output.endswith('qunu> '):
     #   output += shell.recv(4096).decode()
    #print(output)

    # Exit the Telnet session
    shell.send('exit\n')
    time.sleep(1)

    # Wait for the command prompt
    output = ''
    #while not output.endswith('# '):
    output += shell.recv(4096).decode()
    print(output)
    
def CRO():
    # Set up the VISA connection with CRO
    rm = pyvisa.ResourceManager()
    resource_name = 'TCPIP::192.168.10.183::INSTR'  # Replace with your CRO's IP address
    scope = rm.open_resource(resource_name)
    #print(f"Connected to {resource_name}")
    
    #peak to peak voltage measurement
    scope.write('MEASUrement:IMMed:SOUrce1 CH2')
    scope.write('MEASUrement:IMMed:TYPe PK2Pk')
    pktopk_value = scope.query('MEASUrement:IMMed:VALue?')
    
    return(pktopk_value)
```
